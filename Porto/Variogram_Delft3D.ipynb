```python
import os
import time
# import mat73
import h5py
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams.update({'font.size': 12})
plt.rcParams.update({'font.style': 'oblique'})
# data_path = "/Users/yaoling/OneDrive - NTNU/MASCOT_PhD/Data/Porto/D2_3D_salinity-021.mat"
data_path = "/Users/yaoling/OneDrive - NTNU/MASCOT_PhD/Data/Porto/Delft3D/Delft3D.h5"
os.system("say It will take a long time to import data")
t1 = time.time()
# sal_data = mat73.loadmat(data_path)
data = h5py.File(data_path, 'r')
lat = np.array(data.get("lat"))
lon = np.array(data.get("lon"))
depth = np.array(data.get("depth"))
salinity = np.array(data.get("salinity"))
t2 = time.time()
os.system('say Finished Data importing')
print("Time consumed: ", t2 - t1)
print("Everything works fine")
```

```python
def filterNaN(val):
    temp = []
    for i in range(len(val)):
        if not np.isnan(val[i]):
            temp.append(val[i])
    val = np.array(temp).reshape(-1, 1)
    return val

# plot the trend
def plotf(lat, lon, color, title, string):
    plt.scatter(lon, lat, c = color, cmap = "Paired")
    plt.title(title)
    plt.xlabel("Lon [deg]")
    plt.ylabel("Lat [deg]")
    plt.colorbar()
    figpath = "/Users/yaoling/OneDrive - NTNU/MASCOT_PhD/Missions/Porto/Delft3D/fig/"
    plt.savefig(figpath + string + ".pdf")
    plt.show()
    
print(lat.shape)
print(lon.shape)
print(depth.shape)
print(salinity.shape)
plotf(lat[:, :, 0], lon[:, :, 0], np.mean(salinity[:, :, :, 0], axis = 0), "mean of sal", "mean")
# filter data to remove NaN
# lon = sal_data['data']["X"][:, :, 0]
# lat = sal_data['data']['Y'][:, :, 0]
# depth = sal_data['data']['Z'][0, :, :, 0]
# timestamp = (sal_data['data']['Time'] - 719529) * 24 * 3600

# filtered data
# lon_f = filterNaN(lon.reshape(-1, 1))
# lat_f = filterNaN(lat.reshape(-1, 1))
```

```python
# Randomly sample one frame to compare the residual and to find the correlation range
lat_f = lat[:, :, 0].reshape(-1, 1)
lon_f = lon[:, :, 0].reshape(-1, 1)
S_f = np.mean(salinity[:, :, :, 0], axis = 0).reshape(-1, 1)
ind_sample = np.random.randint(0, salinity.shape[0])
S_sample = salinity[ind_sample, :, :, 0].reshape(-1, 1)
# S_sample = filterNaN(sal_data['data']['Val'][ind_sample, :, :, 0].reshape(-1, 1))
residual = S_sample - S_f

print(ind_sample)
plotf(lat_f, lon_f, S_sample, "Sampled salinity field on randomly sampled field {:d}".format(ind_sample), "sample")
plotf(lat_f, lon_f, residual, "Residual of the field", "residual")
```

```python
# Convert lat lon to x y, where x is pointing north, y is pointing east, and depth is z
lat_origin, lon_origin = 41.10251, -8.669811
circumference = 40075000
def deg2rad(deg):
    return deg / 180 * np.pi
def rad2deg(rad):
    return rad / np.pi * 180
def latlon2xy(lat, lon, lat_origin, lon_origin):
    x = deg2rad(lat - lat_origin) / 2 / np.pi * circumference
    y = deg2rad(lon - lon_origin) / 2 / np.pi * circumference * np.cos(deg2rad(lat))
    # x_, y_ = self.R.T @ np.vstack(x, y) # convert it back
    return x, y
x, y = latlon2xy(lat_f, lon_f, lat_origin, lon_origin)

ind = np.random.randint(0, x.shape[0] - 1, size = 5000) # take out only 5000 random locations, otherwise it takes too long to run

range_coef = []
sill_coef = []
nugget_coef = []
# compute the variogram
from progress.bar import Bar
bar = Bar('Processing', max=salinity.shape[0])

from skgstat import Variogram
t1 = time.time()
for i in range(salinity.shape[0]):
    S_sample = salinity[i, :, :, 0].reshape(-1, 1) # sample each frame
    residual = S_sample - S_f
    V_v = Variogram(coordinates = np.hstack((x[ind], y[ind])), values = residual[ind].squeeze(), n_lags = 20, maxlag = 3000, use_nugget=True)
    range_coef.append(V_v.cof[0])
    sill_coef.append(V_v.cof[1])
    nugget_coef.append(V_v.cof[2])
    bar.next()
t2 = time.time()
print("Time consumed: ", t2 - t1)
# bar.finish()
#     print(i)
# V_v.fit_method = 'trf' # moment method
fig = V_v.plot(hist = True)
fig.savefig("test1.pdf")
print(V_v)

plt.figure()
plotf(lat_f[ind], lon_f[ind], S_f[ind], "Sampled trend", "trend")
plt.figure()
plotf(lat_f[ind], lon_f[ind], S_sample[ind], "Sampled sample", "sampled_random")
plt.figure()
plotf(lat_f[ind], lon_f[ind], residual[ind], "Sampled residual", "residual")
```

```python
import seaborn as sns
sns.displot(range_coef, kind = "kde")
plt.axvline(np.mean(range_coef), c = 'r', label = "Mean of range coef is {:.2f}".format(np.mean(range_coef)))
plt.axvline(600, c = 'g', label = "range coef is 600")
plt.legend()
plt.show()

sns.displot(sill_coef, kind = "kde")
plt.axvline(np.mean(sill_coef), c = 'r', label = "Mean of sill coef is {:.2f}".format(np.mean(sill_coef)))
plt.legend()
plt.show()

sns.displot(nugget_coef, kind = "kde")
plt.axvline(np.mean(nugget_coef), c = 'r', label = "Mean of nugget coef is {:.2f}".format(np.mean(nugget_coef)))
plt.legend()
plt.show()
```

```python
# Here comes the depth variogram
range_coef_d = []
sill_coef_d = []
nugget_coef_d = []
# compute the variogram
from progress.bar import Bar
bar = Bar('Processing', max=salinity.shape[0])

S_d_mean = np.mean(salinity, axis = 0)

from skgstat import Variogram
t1 = time.time()
for i in range(salinity.shape[0]):
#     print(i)
    for j in range(50):
#         print(i, j)
        row_ind = np.random.randint(0, salinity.shape[1] - 1)
        col_ind = np.random.randint(0, salinity.shape[2] - 1)
#         print(row_ind, col_ind)
        
        S_sample_d = salinity[i, row_ind, col_ind, :].reshape(-1, 1) # sample each frame
        residual_d = S_sample_d - S_d_mean[row_ind, col_ind, :].reshape(-1, 1)
        Dd = - depth[i, row_ind, col_ind, :].reshape(-1, 1)
        if np.any(np.isnan(Dd)):
            continue
        if len(np.unique(residual_d)) == 1:
            continue
        xx = np.ones_like(Dd).reshape(-1, 1)
#         print(xx)
#         print(Dd)
#         print(residual_d)
#         print(xx.shape)
#         print(Dd.shape)
#         print(residual_d.shape)
        V_v = Variogram(coordinates = np.hstack((xx, Dd)), values = residual_d.squeeze())
        range_coef_d.append(V_v.cof[0])
        sill_coef_d.append(V_v.cof[1])
#         nugget_coef_d.append(V_v.cof[2])

t2 = time.time()
print("Time consumed: ", t2 - t1)
# bar.finish()
#     print(i)
# V_v.fit_method = 'trf' # moment method
fig = V_v.plot(hist = True)
fig.savefig("test1.pdf")
print(V_v)

plt.figure()
plotf(depth[i, row_ind, col_ind, :].reshape(-1, 1), np.zeros_like(residual), S_d_mean[row_ind, col_ind, :], "Sampled trend along depth", "trend_depth")
plt.figure()
plotf(depth[i, row_ind, col_ind, :].reshape(-1, 1), np.zeros_like(residual), salinity[i, row_ind, col_ind, :], "random sample", "sample_depth")
plt.figure()
plotf(depth[i, row_ind, col_ind, :].reshape(-1, 1), np.zeros_like(residual), residual_d, "residual along depth", "residual_depth")
```
